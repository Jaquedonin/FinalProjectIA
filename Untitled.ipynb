```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

enc_letra = LabelEncoder()
enc_menu = LabelEncoder()
enc_fundo = LabelEncoder()
enc_resp = LabelEncoder()
enc_icons = LabelEncoder()
enc_oculos = LabelEncoder()
enc_disp = LabelEncoder()
enc_site = LabelEncoder()

dataset['letra_encoded'] = enc_letra.fit_transform(dataset.iloc[:,0].astype(str))
dataset['menu_encoded'] = enc_menu.fit_transform(dataset.iloc[:,1].astype(str))
dataset['fundo_encoded'] = enc_fundo.fit_transform(dataset.iloc[:,2].astype(str))
dataset['resp_encoded'] = enc_resp.fit_transform(dataset.iloc[:,3].astype(str))
dataset['icons_encoded'] = enc_icons.fit_transform(dataset.iloc[:,4].astype(str))
dataset['oculos_encoded'] = enc_oculos.fit_transform(dataset.iloc[:,6].astype(str))
dataset['disp_encoded'] = enc_disp.fit_transform(dataset.iloc[:,7].astype(str))
dataset['site_encoded'] = enc_site.fit_transform(dataset.iloc[:,8].astype(str))

dataset = dataset.drop(['Tamanho_letra','N_sub_menus','Cor_fundo','Responsividade','Icones','Oculos','Dispositivo', 'Site'], axis=1)
dataset
```

## Teste para verificar se o modelo estÃ¡ correto

Testando o modelo com a prÃ³rpia base usando a coluna que representa cor de fundo 

```python
X_train = dataset.drop(["fundo_encoded"], axis=1)
y_train = dataset["fundo_encoded"]
x_test =  dataset.drop(["fundo_encoded"], axis=1)

neigh = KNeighborsRegressor(n_neighbors=2)

neigh.fit(X_train, y_train)
y_predict = neigh.predict(x_test)
y_predict 
```
